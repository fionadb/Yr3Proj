```python
# Load required libraries
import pandas as pd
```

```python
# Load training data
pdf = pd.read_parquet("../Entire UK data/2013_demand.snappy.parquet")
```

```python
pdf.head()
```

```python
pdf.SETTLEMENT_PERIOD = pd.to_datetime(pdf.SETTLEMENT_PERIOD*1800, unit='s')
pdf['SETTLEMENT_PERIOD'] = [d.time() for d in pdf['SETTLEMENT_PERIOD']]
```

```python
pdf.SETTLEMENT_DATE = pd.to_datetime(pdf.SETTLEMENT_DATE)
pdf['SETTLEMENT_DATE'] = [d.date() for d in pdf['SETTLEMENT_DATE']]
```

```python
pdf.head()
```

```python
pdf['datetime'] = pd.to_datetime(pdf['SETTLEMENT_DATE'].astype(str) + " " + pdf['SETTLEMENT_PERIOD'].astype(str))
pdf['datetime'].head()
```

```python
pdf.columns
```

```python
pdf.drop(['SETTLEMENT_DATE', 'SETTLEMENT_PERIOD', 'ND', 'TSD', 'EMBEDDED_WIND_GENERATION',
       'EMBEDDED_WIND_CAPACITY', 'EMBEDDED_SOLAR_GENERATION',
       'EMBEDDED_SOLAR_CAPACITY', 'NON_BM_STOR', 'PUMP_STORAGE_PUMPING',
       'IFA_FLOW', 'IFA2_FLOW', 'BRITNED_FLOW', 'MOYLE_FLOW', 'EAST_WEST_FLOW',
       'NEMO_FLOW'], axis=1, inplace=True)
```

```python
pdf.head()
```

```python
import pyarrow as pa
import pyarrow.parquet as pq
table = pa.Table.from_pandas(pdf)
pq.write_table(table, '2013_demand_datetime.snappy.parquet')
```
